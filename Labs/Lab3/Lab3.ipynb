```python
import matplotlib.pyplot as plt
import numpy as np

# a) -> b)

x_side = []
y_side = []
x_y = []


with open("unlabelled_data.csv", "r") as unlabelled_data:
     for each_unlabelled_data in unlabelled_data:
         splited_each_unlabelled_data = each_unlabelled_data.split(",")
         x_y.append(tuple(splited_each_unlabelled_data))
         x_side.append(float(splited_each_unlabelled_data[0]))
         y_side.append(float(splited_each_unlabelled_data[1]))


numpy_x_side = np.array(x_side)
numpy_y_side = np.array(y_side)


k, m = np.polyfit(x_side, y_side, 1)


plt.figure(figsize=(10, 6))        
plt.scatter(x_side, y_side, color="orange", label="Unlabbeled Data")
plt.plot(numpy_x_side, k*numpy_x_side+m, lw=0.5)
plt.legend()


plt.xlabel("x-axes")        
plt.ylabel("y-axes")   
plt.title("Unlabbelled data")


x_values = tuple(range(round(min(x_side)), round(max(x_side))+1))
y_values = tuple(range(round(min(y_side)), round(max(y_side))+1))
plt.xticks(x_values)
plt.yticks(y_values)


plt.grid()
plt.axhline(0, color="black", lw=0.5)
plt.axvline(0, color="black", lw=0.5)


plt.show()

```
